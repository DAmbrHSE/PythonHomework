```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
def GetDocumentOwner(doc_ID):
    """ get document owner name from document ID """
    for document in documents:
        if document['number'] == doc_ID:
            return document['name']
    
    return None # there's no document with this ID]
```

```python
def GetDocumentPlacement(doc_ID):
    """ get document directory index from document ID """
    for directory_index in directories:
        for document in directories[directory_index]:
            if document == doc_ID:
                return directory_index
    
    return None # there's no document with this ID
```

```python
while True:
    print("Введите команду:")
    command = input()
    
    if (command == "p"):
        print("Введите номер документа:")
        doc_ID = input()
        
        doc_owner = GetDocumentOwner(doc_ID)
        if doc_owner == None:
            print("Такого документа не существует.")
        else:
            print("Владелец документа " + doc_ID + ": " + doc_owner)
    if (command == "s"):
        print("Введите номер документа:")
        doc_ID = input()
        
        doc_placement = GetDocumentPlacement(doc_ID)
        if doc_placement == None:
            print("Такого документа не существует.")
        else:
            print("Документ " + doc_ID + " хранится на полке: " + doc_placement)
    elif (command == "q"):
        print("Конец работы программы.")
        break
```
