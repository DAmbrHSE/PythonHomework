```python
import json

# читать purchase log в оперативную память не запрещено заданием
f_purchase_log = open('purchase_log.txt', 'r')
purchaces = {}

for line in f_purchase_log:
    dict = json.loads(line.strip())
    purchaces[dict['user_id']] = dict['category']

f_purchase_log.close()
```

```python
f_visit_log = open('visit_log.csv', 'r')
f_out = open('funnel.csv', 'w')

for line in f_visit_log:
    line_content = line.strip().split(',')
    category = purchaces.get(line_content[0])

    if (category != None):
        f_out.write(line.strip() + ","+ category + "\n")

f_visit_log.close()
f_out.close()
```
