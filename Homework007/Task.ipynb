```python
def ParseCSVTable( file_name ):
    try:
        table_file = open(file_name, 'r')
    except FileNotFoundError:
        return None

    names = table_file.readline().strip().split(',')
    data = []

    for data_line_string in table_file:
        data_line = data_line_string.strip().split(',')
        data.append(dict(zip(names, data_line)))

    table_file.close()
    return data
```

```python
def FormatInfoToString( info ):
    string = '{name}, {sex} {age}-ти лет, находившись в {region}, с {device}а {have_purchased} на {bill} у.е. в браузере {browser}.'

    sex_names = {'female' : 'женщина', 'male' : 'мужчина'}
    have_purchased = { 'female' : 'совершила покупку', 'male' : "совершил покупку"}
    device_names = {'mobile' : "телефон", 'tablet' : 'планшет', 'laptop' : 'ноутбук', 'desktop' : 'компьютер'}
    region_str = info['region'] if (info['region'] != '-') else 'неизвестном регионе'

    return string.format(name=info['name'], sex=sex_names[info['sex']], age=info['age'], region=region_str, device=device_names[info['device_type']],
                        have_purchased=have_purchased[info['sex']], bill=info['bill'], browser=info['browser'])
```

```python
def PrintWebInfoToTXT( data, file_name ):
    try:
        out_file = open(file_name, 'w', encoding="utf-8")
    except:
        return
    
    for data_line in data:
        print(FormatInfoToString(data_line), file=out_file)
    out_file.close()
    return
```

```python
data = ParseCSVTable('web_clients_correct.csv')
PrintWebInfoToTXT(data, 'web_clients_result.txt')
```
